```python
import pandas as pd
from astral import LocationInfo
from astral.sun import sun

df = pd.read_csv("mapdata.csv", sep=';')

# Parse timestamps as local Finland time
df['datetime'] = pd.to_datetime(df['datetime'])

df['datetime'] = df['datetime'].dt.tz_localize('Europe/Helsinki')

def get_light(row):
    observer = LocationInfo(latitude=row['lat'], longitude=row['lon'])
    
    s = sun(observer.observer, date=row['datetime'].date(), tzinfo=row['datetime'].tzinfo)

    t = row['datetime']

    if s['sunrise'] <= t <= s['sunset']:
        return "Day"
    elif s['dawn'] <= t < s['sunrise']:
        return "Dawn"
    elif s['sunset'] < t <= s['dusk']:
        return "Dusk"
    else:
        return "Night"

df['light'] = df.apply(get_light, axis=1)
```

```python
df.drop(columns=["is_dark"], inplace=True)
```
